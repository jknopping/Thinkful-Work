```python
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

### Method 1: Identifying positive reviews

```python
#Choosing amazon to start
data_path = ("amazon_cells_labelled.txt")

sms_raw = pd.read_csv(data_path, delimiter= '\t', header=None)
#1 is positive, 0 is negative
sms_raw.columns = ['message', 'positive']
```

```python
#Find all positive messages
print(sms_raw.loc[sms_raw.positive == 1].count())
sms_raw.loc[sms_raw.positive == 1]
```

```python
#Create wordcloud of postive reviews
from wordcloud import WordCloud

wordcloud = WordCloud(background_color='black').generate(" ".join(sms_raw.loc[sms_raw.positive == 1].message))
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()

#sms_raw.loc[sms_raw.positive == 1]
```

```python
#Create wordcloud of negative reviews
from wordcloud import WordCloud

wordcloud = WordCloud(background_color='orange').generate(" ".join(sms_raw.loc[sms_raw.positive == 0].message))
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()
```

```python
#use keywords obtained from wordcloud that are found mostly with posistive reviews
keywords = ['great',
            'good',
            'well',
            'excellent',
            'best',
            'love',
            'happy',
            'nice',
            'quality',
            'sturdy',
            'worked',
            'clear',
            'comfortable']

for key in keywords:
    sms_raw[str(key)] = sms_raw.message.str.contains(" " + str(key) + " ", case=True)
    sms_raw[str(key)] = sms_raw.message.str.contains(str(key) + " ", case=True)
    sms_raw[str(key)] = sms_raw.message.str.contains(str(key) + ".", case=True)
    
```

```python
sms_raw['positive'] = (sms_raw['positive'] == 1)
```

```python
sns.heatmap(sms_raw.corr())
```

```python
data = sms_raw[keywords]

target = sms_raw['positive']

# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

#Display our results
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))

accuracy = 100*((target == y_pred).sum())/data.shape[0]
print('Accuracy: {}%'.format(accuracy))
```

```python
def calculate_cm(y_pred, target):
    tp = 0 #True Positives
    tn = 0 #True Negatives 
    fp = 0 #False Positives
    fn = 0 #False Negatives
    x = 0
    while x < len(y_pred):
        if y_pred[x] == 0 and target[x] == 0:
            tn += 1
        if y_pred[x] == 0 and target[x] == 1:
            fp += 1
        if y_pred[x] == 1 and target[x] == 1:
            tp += 1
        if y_pred[x] == 1 and target[x] == 0:
            fn += 1
        x += 1
    cm = [[tn, fp],[fn, tp]]
    sens = (tp)/(tp + fp) *100
    spec = tn/(tn + fn) *100
    print('Sensitivity: {}%'.format(sens))
    print('Specificity: {}%'.format(spec))
    return cm

calculate_cm(y_pred, target)
```

```python
#Holdout groups
from sklearn.model_selection import train_test_split
# Use train_test_split to create the necessary training and test groups
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=20)
print('With 20% Holdout: ' + str(bnb.fit(X_train, y_train).score(X_test, y_test)))
print('Testing on Sample: ' + str(bnb.fit(data, target).score(data, target)))
```

```python
#Cross validation
from sklearn.model_selection import cross_val_score
cross_val_score(bnb, data, target, cv=10)
```

Method 1 has a low sensitivity at 44% with 280 false positives.

However, Method 1 has high specificity at 89.2%, with only 52 false negatives.
